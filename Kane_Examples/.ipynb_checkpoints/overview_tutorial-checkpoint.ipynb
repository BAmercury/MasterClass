Kane's Method Roadmap:
- Define generalized coordinates/speeds
- For each rigid body B, find w_b_n and v_b_n
- Find partial velocities for each degree of freedom (v_r_n_b, w_r_n_b)
- Find expression for acceleration for each rigid body
- Find active torques/forces acting on each body
- Assemble Newton's/Euler's equations (F-ma = 0) for each rigid body
- For each generalized speed, project into its direction and sum

###  Example 1: Square plate suspended from a fixed support by the means of 2 springs and two adjacent vertices

One of the strings is cut instantaneously without any change in the configurations of the other components in the system

The objective is to simulate the motion that the plate will exhibit when one of the strings is cut

![Example 1 Image](imgs/example1.PNG)


Some notes about the system:
- Tension in the string will always be positive and string will remain taut (Can treat string as a rigid link)
- Will assume this as a massless link/string
- Joints considered frictionless


```python
# Define constants
L = symbols('L') # Length of string
a = symbols('a') # Length of the block
# Describe kinematic variables
# Angles and their derivatives
theta, phi = dynamicsymbols('theta phi') 
theta_d, phi_d = dynamicsymbols('theta phi', 1)
# Generalized speeds
u1, u2 = dynamicsymbols('u1 u2')
u1d, u2d = dynamicsymbols('u1 u2', 1)

# Define reference frames
# Inertial
Nframe = ReferenceFrame('Nframe')
# Define frame attached to point G
Gframe = Nframe.orientnew('Gframe', 'Axis', [theta, Nframe.z])
# Define frame attached to point C
Cframe = Gframe.orientnew('Cframe', 'Axis', [phi, Nframe.z])

# set angular velocity of these frames 

# w_g/n = theta_d * N.z
Gframe.set_ang_vel(Nframe, u1 * Nframe.z)
Cframe.set_ang_vel(Nframe, u2 * Nframe.z)

# Define locations
O = Point('O')
# Point G wrt Point O
G = O.locatenew('G', -L/2 * Gframe.y)
# Point P wrt Point G
P = G.locatenew('P', -L/2 * Gframe.y)
# Point C wrt to point P
C = P.locatenew('C', a/2 * (Cframe.x - Cframe.y))
# Point R wrt to point P
R = P.locatenew('R', -a*Cframe.y)

# Now define point velocities 
O.set_vel(Nframe, 0)
# Set other velocities
G.v2pt_theory(O, Nframe, Gframe)
P.v2pt_theory(O, Nframe, Gframe) # Point P is connected to point O via Gframe
C.v2pt_theory(P, Nframe, Cframe) # Point C moves with respect to point P connected to point P via the Pframe
R.v2pt_theory(P, Nframe, Cframe) # Point R moves wrt point P connected to point P via the Pframe

# Gather everything into lists
gen_coordinates = [theta, phi]
gen_speeds = [u1, u2]

# form kinematic differential equations
kde = [u1 - theta_d, u2 - phi_d]

# Initialize Kane's object
kane = KanesMethod(Nframe, gen_coordinates, gen_speeds, kde)

# Inertia dynamics
Ig, Ic = symbols('Ig Ig') # Inertia of string and block

mg, mc = symbols('mg, mc') # Mass of link and block

g = symbols('g') # gravitational constant

itg = inertia(Gframe, 0, 0, Ig) # In G body fixed frame
ig_tuple = (itg, G) # Defined at G point

itc = inertia(Cframe, 0, 0, Ic) # In C body fixed frame
ic_tuple = (itc, C) # Defined at C point

# Define rigid body objects
link = RigidBody('Link', G, Gframe, mg, ig_tuple)
block = RigidBody('Block', C, Cframe, mc, ic_tuple)

# Define forces and point of application

# Gravity on each rigid body
f_g_g = (G, -mg*g*Nframe.y) # Applied at point G
f_g_c = (C, -mc*g*Nframe.y) # Applied at point C
loads = [f_g_g, f_g_c]
bodies = [link, block]

# Get equations
fr, frstar = kane.kanes_equations(bodies, loads)
trigsimp(fr+frstar)



```

```python
# Obtain mass matrix and forcing vector
mm = trigsimp(kane.mass_matrix_full)
fm = trigsimp(kane.forcing_full)

# List of all constants
constants = [L, mg, Ic, a, mc, Ic, g]
num_constants = array([12, 0, 0, 6, 10, 60, 9,81])

# Generate RHS
rhs_eqns = generate_ode_function(fm, gen_coordinates, gen_speeds, constants, mass_matrix = mm)

# Define initial conditions
x0 = zeros(4)
# Time
t = linspace(0, 50, 1000)

rhs_eqns(x0, 0.0, num_constants)

y = odeint(rhs_eqns, x0, t, args=(num_constants, ))

plt.plot(t, rad2deg(y[:,:2])) # first two columns of y contain theta and phi
plt.xlabel('Time [sec]')
plt.ylabel('Angle [deg]')
plt.legend(["{}".format(vlatex(c)) for c in gen_coordinates])
```

```python
# 3D Vs
from pydy.viz.shapes import Cylinder, Cube
import pydy.viz
from pydy.viz.visualization_frame import VisualizationFrame
import pydy.viz.scene
from pydy.viz.scene import Scene
import pythreejs

constants_dict = dict(zip(constants, num_constants))

link_shape = Cylinder(length=constants_dict[L], radius = 0.1, color = 'gray')
block_shape = Cube(length=constants_dict[a], color = 'blue')
lvf = VisualizationFrame('Link', Gframe, G, link_shape)
cvf = VisualizationFrame('Block', Cframe, C, block_shape)

scene = Scene(Nframe, O)
scene.visualization_frames = [lvf, cvf]
scene.state_symbols = gen_coordinates + gen_speeds
scene.constants = constants_dict
scene.states_trajectories = y

#scene.display_jupyter(axes_arrow_length=1.0)
scene.display()
#scene.display_jupyter(axes_arrow_length=1.0)
```
